```julia
using Plots, DelimitedFiles, FileIO, LaTeXStrings
```

## Read volumes

```julia
# Leo volumen de la traj
lv_trj_vol_1 = convert(Array{Float64, 1}, readdlm("1_lv_volume")[:, 3])
hv_trj_vol_1 = convert(Array{Float64, 1}, readdlm("1_hv_volume")[:, 3])

lv_trj_vol_2 = convert(Array{Float64, 1}, readdlm("2_lv_volume")[:, 3])
hv_trj_vol_2 = convert(Array{Float64, 1}, readdlm("2_hv_volume")[:, 3])

lv_trj_vol_3 = convert(Array{Float64, 1}, readdlm("3_lv_volume")[:, 3])
hv_trj_vol_3 = convert(Array{Float64, 1}, readdlm("3_hv_volume")[:, 3]);
```

## Full trajectory, without ligand

```julia
plot_1 = histogram(lv_trj_vol_1, color = "blue",
    bins = collect(0:50:1300), label = "Low Volume",
    yaxis = "Count", xaxis = "Volume A3")

histogram!(plot_1, hv_trj_vol_1, color = "crimson", 
    bins = collect(0:50:1300), label = "High Volume")
```

## Full trajectory with and without ligand

```julia
plot_2 = histogram(lv_trj_vol_1, color = "blue",
    bins = collect(0:50:1300), label = "Low Volume",
    yaxis = "Count", xaxis = "Volume A3")

histogram!(plot_2, lv_trj_vol_2, color = "lightskyblue",
    bins = collect(0:50:1300), label = "Low Volume w/ PLM")
```

```julia
plot_3 = histogram(hv_trj_vol_1, color = "crimson", 
    bins = collect(0:50:1300), label = "High Volume",
    yaxis = "Count", xaxis = "Volume A3")

histogram!(plot_3, hv_trj_vol_2,color = "tomato", 
    bins = collect(0:50:1300), label = "High Volume w/ PLM")
```

### Saving plots

```julia
savefig(plot_1, "plot_1.png")
savefig(plot_2, "plot_2.png")
savefig(plot_3, "plot_3.png")
```
